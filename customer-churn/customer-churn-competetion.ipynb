```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.ensemble import RandomForestClassifier,GradientBoostingRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
df_train = pd.read_csv('/kaggle/input/playground-series-s4e1/train.csv')
```

```python
df_train = pd.get_dummies(df_train,columns=['Geography','Gender'])
```

```python
X = df_train.drop(columns=['id','Exited','CustomerId','Surname'])
y = df_train['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.25,random_state=34)
```

```python
rf_model = RandomForestClassifier(n_estimators=100)
# rf_model.fit(x_train,y_train)
# y_pred = rf_model.predict(x_test)
# print(f"ROC AUC Score: {roc_auc_score(y_test,y_pred)}")
```

```python
param_grid = {
    'n_estimators':[80,90,100,110,120]
}
grid = GridSearchCV(rf_model,param_grid=param_grid,cv=5,n_jobs=-1,scoring = 'roc_auc')
grid.fit(X,y)
```

```python
gb_model = GradientBoostingRegressor()
param_grid2 = {
    "n_estimators":[80,90,100,110,120],
    "max_depth": [5,9,13,14]
}
grid2 = GridSearchCV(gb_model,param_grid=param_grid2,cv=5,n_jobs=-1,scoring = 'roc_auc')
grid2.fit(X,y)
```

```python
df_test = pd.read_csv('/kaggle/input/playground-series-s4e1/test.csv')
df_test = pd.get_dummies(df_test,columns=['Geography','Gender'])
id_ = df_test['id'] 
df_test = df_test.drop(columns=['id','CustomerId','Surname'])
y_pred = grid.best_estimator_.predict(df_test)
```

```python
y_pred = pd.Series(y_pred)
```

```python
df_final = pd.concat([id_,y_pred],axis = 1)
df_final.to_csv('fresult.csv',index=False)
df_final.head()
```
