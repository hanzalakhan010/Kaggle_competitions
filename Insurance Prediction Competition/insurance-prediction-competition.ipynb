```python
#dropping columns with different datatypes
to_drop = ['id','Customer Feedback','Previous Claims','Policy Start Date']
df_train.drop(columns = to_drop,inplace = True)
```

```python
to_fillna_median = ['Credit Score',"Health Score",'Number of Dependents','Annual Income',"Age","Insurance Duration",'Vehicle Age']
to_fillna_mode = ['Occupation','Marital Status']

df_train[to_fillna_median] = df_train[to_fillna_median].fillna('median')
df_train[to_fillna_mode] = df_train[to_fillna_mode].fillna('mode')
```

```python
X = df_train.drop(columns = ['Premium Amount'])
y = df_train['Premium Amount']
```

```python
gb_model = GradientBoostingRegressor()
param_grid = {
    'n_estimators':[80,90,100,110],
    "max_depth":[10,15,20]
}
grid = GridSearchCV(gb_model,param_grid = param_grid,cv = 4,n_jobs = -1)
```
